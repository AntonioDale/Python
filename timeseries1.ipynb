```python
# Importo il dataset Umidità
umidita = pd.read_csv(r"C:\Users\39340\Desktop\PY_Challenge\#16 Time Series (1)\humidity.csv", parse_dates = ["datetime"])
umidita.set_index("datetime", inplace = True)
```

```python
# Identifichiamo i nulli
umidita.isna().sum()
umidita.shape
nulli = umidita.isna().sum().reset_index()
nulli["%"] = round(nulli[0] / umidita.shape[0]*100,2)
nulli
```

```python
# Correzione dei nulli
umidita.fillna(method = "ffill", inplace = True)
umidita.fillna(method = "bfill", inplace = True)
```

```python
umidita.isna().sum().reset_index()
```

```python
# Convertiamo il passo dei dataframe
umidita.asfreq("M") #Mensile
```

```python
umidita.asfreq("W") # Settimanale
```

```python
umidita.asfreq("Y") # Annuale
```

```python
# Plotting Giornaliero
umi_day = umidita.asfreq("D")
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        y = umi_day["San Francisco"],
        x = umi_day.index,
        mode = "lines+markers",
        line_color = "#7242f5"
    )
)

fig.update_layout(
    template = "plotly_white",
    font_color = "black",
    height = 500,
    width = 1000
)
```

```python
# Settimanale
umi_set = umidita.asfreq("W")
umi_med_set = umidita["San Francisco"].resample("W").agg("mean").reset_index()

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        y = umi_set["San Francisco"],
        x = umi_set.index,
        mode = "lines+markers",
        line_color = "#7242f5",
        line_width = 1.5,
        marker_size = 4
    )
)
fig.add_trace(
    go.Scatter(
        y = umi_med_set["San Francisco"],
        x = umi_set.index,
        mode = "lines+markers",
        line_color = "#f54266",
        line_width = 0.8,
        marker_size = 3
    )
)

fig.update_layout(
    template = "plotly_white",
    font_color = "black",
    height = 500,
    width = 1000
)
```

```python
# Media della settimana
umi_med_set = umidita["San Francisco"].resample("W").agg("mean").reset_index()
umi_med_set
```

```python
# Mensile
umi_men = umidita.asfreq("M")
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        y = umi_men["San Francisco"],
        x = umi_men.index,
        mode = "lines+markers",
        line_color = "#7242f5"
    )
)

fig.update_layout(
    template = "plotly_white",
    font_color = "black",
    height = 500,
    width = 1000
)
```

```python
# Annuale
umi_y = umidita.asfreq("Y")
umi_med_ann = umidita["San Francisco"].resample("AS").agg("mean").reset_index()

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        y = umi_y["San Francisco"],
        x = umi_y.index,
        mode = "lines+markers",
        line_color = "#7242f5"
    )
)

fig.add_trace(
    go.Scatter(
        y = umi_med_ann["San Francisco"],
        x = umi_med_ann["datetime"],
        mode = "lines+markers",
        line_color = "blue"
    )
)

fig.update_layout(
    template = "plotly_white",
    font_color = "black",
    height = 500,
    width = 1000
)
```

```python
# Plottiamo i dati di chiusura di google tra 2015 e 2017
google_dat = google["2015":"2017"]

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = google_dat.index,
        y = google_dat["Open"],
        mode = "lines",
        line_color = "#87f542",
        line_width = 3
    )
)

fig.update_layout(
    template = "plotly_white",
    font_color = "black",
    height = 500,
    width = 700
)
```

```python
# Shift e lag
atl_mens = umidita["Atlanta"].asfreq("M").reset_index()
at_shift = umidita["Atlanta"].asfreq("M").shift(periods = 12).reset_index()
at_shift.rename(columns = ({"Atlanta":"shift"}), inplace = True)
atl_mens["shift"] = at_shift["shift"]
atl_mens.head(20)
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x = atl_mens["datetime"],
        y = atl_mens["Atlanta"],
        line_color = "#f5b942"
    )
)

fig.add_trace(
    go.Scatter(
        x = atl_mens["datetime"],
        y = atl_mens["shift"],
        line_color = "#f55442"
    )
)

fig.update_layout(
    template = "plotly_white",
    height = 500,
    width = 800
    )
```

```python
# Rendimento Giornaliero
google["shift_g"] = google["Close"].shift(periods = 1)
google["rend_g"] = round((google["Close"]/google["shift_g"]-1)*100, 2)
google
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x = google.index,
        y = google["rend_g"],
        line_color = "#f5b942"
    )
)

fig.update_layout(
    template = "plotly_white",
    height = 500,
    width = 800
    )
```

```python
# Calcoliamo e plottiamo il rendimento mensile
google_month_mean = google.resample("M").agg("mean")
google_month_mean.drop(["shift_g", "rend_g"], axis = 1, inplace = True)
google_month_mean["shift"] = google_month_mean["Close"].shift(periods = 1)
google_month_mean["rend_m"] = round((google_month_mean["Close"]/google_month_mean["shift"]-1)*100, 2)
google_month_mean
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = google_month_mean.index,
        y = google_month_mean["rend_m"],
        line_color = "#425df5"
    )
)

fig.update_layout(
    template = "plotly_white",
    height = 500,
    width = 800,
    font_color = "black"
    )
```

```python
# Differenza
google_diff = google_month_mean["Close"].diff().reset_index()

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = google_diff["Date"],
        y = google_diff["Close"],
        line_color = "#425df5"
    )
)

fig.update_layout(
    template = "plotly_white",
    height = 500,
    width = 800,
    font_color = "black"
)
```
